```python
from random import randint
```

```python
ranks=[0,'Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
color=[0,'Clubs','Diamond','Hearts','Spades']
track_card=[]
value=[0,11,2,3,4,5,6,7,8,9,10,10,10,10]
```

```python
class pl():
    
    def __init__(self,amount=100,total=0):
        self.cards=[]
        self.amount=amount
        # Initialize cards
        cnt=0
        while cnt<2:
            r = randint(1,13)
            c = randint(1,4)
            card = c*100+r
            if card not in track_card:
                track_card.append(card)
                self.cards.append(card)
                cnt+=1
            else:
                continue
                
            # Initialize total
        for i in self.cards:
            total=total+value[int(i%100)]
        self.total=total
        
    def __str__(self):
        string='player :\n'
        for c in self.cards:
            string = string + ranks[int(c%100)] + ' of ' + color[int(c/100)] + '\n'
        string = string + 'total=' + str(self.total) + '\n'
        return string
    
    def gen_card(self):
        while True:
            r = randint(1,13)
            c = randint(1,4)
            card = c*100+r
            if card not in track_card:
                track_card.append(card)
                self.cards.append(card)
                self.total=self.total+value[int(card%100)]
                break
            else:
                continue
```

```python
class dl():
    
    def __init__(self,total=0):
        self.cards=[]
        self.mode='hidden'
        # Initialize cards
        cnt=0
        while cnt<2:
            r = randint(1,13)
            c = randint(1,4)
            card = c*100+r
            if card not in track_card:
                track_card.append(card)
                self.cards.append(card)
                cnt+=1
            else:
                continue
        # Initialize total
        for i in self.cards:
            total=total+value[int(i%100)]
        self.total=total
                
    def __str__(self):
        string='dealer :\n'
        if self.mode == 'hidden':
            string=string+'<hidden card!>\n'
            for c in range(1,len(self.cards)):
                c1=self.cards[c]
                string = string + ranks[int(c1%100)] + ' of ' + color[int(c1/100)] + '\n'
        else:
            for c in self.cards:
                string = string + ranks[int(c%100)] + ' of ' + color[int(c/100)] + '\n'
            string=string+'total='+str(self.total)+'\n'
        return string
    
    def gen_card(self):
        while True:
            r = randint(1,13)
            c = randint(1,4)
            card = c*100+r
            if card not in track_card:
                track_card.append(card)
                self.cards.append(card)
                self.total=self.total+value[int(card%100)]
                break
            else:
                continue

    
    def play(self):
        while True:
            if self.total<17:
                self.gen_card()
            else:
                self.mode='show'
                break
```

```python
print("WELCOME TO BLACKJACK GAME!\n")
player=pl()
dealer=dl()

while True:
    print(dealer)
    print(player)
    print("Do you want to hit or stay?(h/s)")
    char=input()
    
    if char=='h':
        player.gen_card()
        print(player)
        if player.total>21:
            print("Player busted!")
            break
    elif char=='s':
        dealer.play()
        if dealer.total>21:
            print("Dealer is busted!")
            print("Player wins!")
            break
        print(dealer)
        print(player)
        if dealer.total>player.total:
            print("Dealer Wins!")
            break
        elif dealer.total<player.total:
            print("Player Wins!")
            break
        else:
            print("It`s tie!!!")
            
    else:
        print("Enter char only 'h'/'s'")
        continue
    
```
