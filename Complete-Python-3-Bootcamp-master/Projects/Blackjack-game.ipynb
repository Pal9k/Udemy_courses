```python
from random import randint
```

```python
ranks=[0,'Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
color=[0,'Clubs','Diamond','Hearts','Spades']
track_card=[]
value=[0,11,2,3,4,5,6,7,8,9,10,10,10,10]
```

```python
class gamer():        
    def __init__(self):
        self.cards=[]          # cards having gamer(player/dealer)
        self.total=0           # total having gamer(player/dealer)
        self.ace=[]            # the ace having gamer (which will be used to track when total goes over 21 n having ace)
        # Initialize cards
        self.gen_card()
        self.gen_card()
        
    def gen_card(self):
        while True:
            r = randint(1,13)    # rank
            c = randint(1,4)     # color
            card = c*100+r       # card 
            
            if card not in track_card:     # if card not used either dealer or player
                track_card.append(card)
                self.cards.append(card)
                
                # Initialize total
                self.total=self.total+value[int(card%100)]
                
                # correct for Ace
                if card%100==1:
                    self.ace.append(card)   # track of ace
                self.correct_ace()
                break
            else:
                continue
                
    def correct_ace(self):
        if self.total>21:
            if len(self.ace)>0:
                self.total=self.total-10
                self.ace.pop()
            
        
```

```python
class pl(gamer):
    
    def __init__(self):
        super().__init__()
        self.amount=100
        self.name='Player'
        
    def __str__(self):
        string=self.name+'\n'
        for c in self.cards:
            string = string + ranks[int(c%100)] + ' of ' + color[int(c/100)] + '\n'
        string = string + 'total=' + str(self.total) + '\n'
        return string
```

```python
class dl(gamer):
    
    def __init__(self):
        super().__init__()
        self.name='Dealer'
        self.mode='hidden'
                
    def __str__(self):
        string=self.name+':\n'
        # if card need to hidden
        if self.mode == 'hidden':
            string=string+'<hidden card!>\n'
            for c in range(1,len(self.cards)):
                c1=self.cards[c]
                string = string + ranks[int(c1%100)] + ' of ' + color[int(c1/100)] + '\n'
        # if card need to show
        else:
            for c in self.cards:
                string = string + ranks[int(c%100)] + ' of ' + color[int(c/100)] + '\n'
            string=string+'total='+str(self.total)+'\n'
        return string

    
    def play(self):
        # dealer is playing it`s game usinf following stratagy!
        while True:
            if self.total<17:
                self.gen_card()
            else:
                self.mode='show'
                break
```
